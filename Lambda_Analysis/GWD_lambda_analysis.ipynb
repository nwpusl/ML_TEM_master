```python
import numpy as np
from sklearn.linear_model import LinearRegression
from itertools import combinations

import pandas as pd

# Load the data from the Excel file
data = pd.read_excel('lambda_data.xlsx', sheet_name='GWD_data')

# Fill missing values with 0
data = data.fillna(0)

# Extract target variable ('lambda') and features
data_target = data['lambda']
feature_names = ['cC', 'cO', 'cN', 'cF', 'cP', 'OCr', 'FOr',
                 'NOr', 'PFr', 'NSr']  # Top 10 selected features
data_features_part = data[feature_names]
y = data_target.values

# Define the R^2 threshold for filtering models
r2_threshold = 0.38

# Iterate over all possible combinations of 3 features
for feature_comb in combinations(feature_names, 3):
    # Extract the feature columns for the current combination
    a = data_features_part[feature_comb[0]].values
    b = data_features_part[feature_comb[1]].values
    c = data_features_part[feature_comb[2]].values
    X = np.column_stack((a, b, c))

    # Create an instance of the linear regression model
    model = LinearRegression()  # Uses ordinary least squares by default

    # Fit the model to the data
    results = model.fit(X, y)

    # Calculate the goodness-of-fit metric (R^2)
    r_squared = model.score(X, y)

    # Output the fitted results and R^2 if it exceeds the threshold
    if r_squared > r2_threshold:
        print(f'Feature Combination: {feature_comb}')
        print(f'R^2: {r_squared}')
        print(f'Coefficients: {model.coef_}')
        print(f'Intercept: {model.intercept_}')
```
