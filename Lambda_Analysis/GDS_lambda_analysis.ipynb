```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from itertools import combinations
import pandas as pd

data = pd.read_excel('lambda_data.xlsx', sheet_name='GDS_data')
data = data.fillna(0)
data_target = data['lambda']
feature_names = ['cN','cO','NOr']


# feature_names = ['cO','cN','cF','cS','FCr','FOr',
#                'NFr','NOr','NSr','FP/Or']  #top10
fea_num = len(feature_names)
data_features_part = data[feature_names]

y = data_target.values

r2_threshold = 0.3 


for k in range(3, 4): 
    feature_combinations = combinations(feature_names, k)
    for combination in feature_combinations:
        combination = list(combination)
        X = data_features_part[combination].values

        poly = PolynomialFeatures(degree=4)
        X_poly = poly.fit_transform(X)

        model = LinearRegression()
        
        model.fit(X_poly, y)

        r2 = model.score(X_poly, y)
        
        coefficients = model.coef_

        intercept = model.intercept_
        
        # print(f'Features: {combination}')
        # print(f'R^2: {r2}')
        # print()

        if r2 > r2_threshold:
            print('X_poly.shape:',X_poly.shape)
            print('feature_names:')
            for i in combination:
                print(i,end=' ')
            print()
            print(f'Coefficients: {coefficients}')
            print(f'Intercept: {intercept}')  # intercept
            print(f'R^2: {r2}')
```
